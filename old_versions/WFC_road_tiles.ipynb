```python
import sys
if "newtiles" in sys.modules.keys():
    del sys.modules["newtiles"]
import newtiles as t
import pygame
from typing import List
import numpy as np

try:
    pygame.quit()
except:
    pass

boardSize = 10
screensize = 800
tilesize = screensize // boardSize



pygame.init()
pygame.display.set_caption('Carcarsonne tiles')
screen = pygame.display.set_mode((screensize, screensize))

pygame.event.pump()
pygame.display.update()
pygame.time.wait(100)


# making the tiles

curve_surf = pygame.image.load("roads/curve.png")
straight_surf = pygame.image.load("roads/straight.png")
end_surf = pygame.image.load("roads/end.png")

curve    = t.ImageTile(["grg", "grg", "ggg", "ggg"], image_surface=curve_surf)
straight = t.ImageTile(["ggg", "grg", "ggg", "grg"], image_surface=straight_surf)
end      = t.ImageTile(["grg", "ggg", "ggg", "ggg"], image_surface=end_surf)


curves = [curve, curve.copy().rot90(), curve.copy().rot90().rot90(), curve.copy().rot90().rot90().rot90()]
straights = [straight, straight.copy().rot90()]
ends = [end, end.copy().rot90(), end.copy().rot90().rot90(), end.copy().rot90().rot90().rot90()]

all = curves + straights + ends


# initializing the manifold with all tiles possible

manifold = t.Manifold(boardSize, boardSize)
for x in range(boardSize):
    for y in range(boardSize):
        manifold.collections.append(t.TileCollection(all))

# manifold.collections[y*boardSize + x].tiles = [straight,]


def showBoard(speed_ms : int = 100):
    
    screen.fill((100, 100, 100))

    # draw grid
    for x in range(1, boardSize):
        pygame.draw.line(screen, (0, 0, 0), (x * tilesize, 0), (x * tilesize, screensize), width=3)
        pygame.draw.line(screen, (0, 0, 0), (0, x * tilesize), (screensize, x * tilesize), width=3)
        

    # draw all the tiles
    for x in range(boardSize):
        for y in range(boardSize):
            
            manifold.collections[y * boardSize + x].draw(screen, np.array([x, y]), tilesize)
            pygame.event.pump()
            pygame.display.update()
            pygame.time.wait(speed_ms // (boardSize * boardSize))

def delay(ms : int):
    pygame.time.wait(ms)
    pygame.event.pump()
    pygame.display.update()

showBoard(speed_ms=200)


while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_ESCAPE:
                pygame.quit()
            if event.key == pygame.K_SPACE:
                manifold = t.Manifold.collapseRandom(manifold)
                showBoard(speed_ms=0)
                

```
